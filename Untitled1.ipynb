```python
import pandas as pd

# Create simple dataset
data = {
    "age": [25, 45, 50, 35, 60, 40, 55, 30],
    "cholesterol": [180, 220, 240, 200, 260, 210, 250, 190],
    "bp": [120, 140, 150, 130, 160, 135, 155, 125],
    "disease": [0, 1, 1, 0, 1, 0, 1, 0]  # 0 = No, 1 = Yes
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.model_selection import train_test_split

# Features (input)
X = df[["age", "cholesterol", "bp"]]

# Target (output)
y = df["disease"]

# Split into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("Training data:", len(X_train))
print("Testing data:", len(X_test))
```

```python
print(X_train)
```

```python
from sklearn.linear_model import LogisticRegression

# Create model
model = LogisticRegression()

# Train model using training data
model.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
# Use the model to make predictions on the test set
y_pred = model.predict(X_test)

# Compare the predictions to the actual results
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)

print(f"Predictions: {y_pred}")
print(f"Actual Labels: {y_test.values}")
print(f"Model Accuracy: {accuracy * 100}%")
```
